### Plotting Kernel Density Estimation

```python
def KDE(values, x, bandwidth, ymax=None):
    kde = KernelDensity(kernel='gaussian', bandwidth=bandwidth).fit(values)
    dens = kde.score_samples(x)
    mi_ind = argrelextrema(dens, np.less)[0]
    mi_value = x[mi_ind]
    print('number of groups: {}'.format(mi_value.size))
    if (ymax == None):
        ymax = np.exp(dens.max())
    
    plt.figure(figsize=(20,10))
    plt.plot(x, np.exp(dens), '-')
    plt.plot(values, -(ymax*0.02)-(ymax*0.02)*np.random.random(values.shape[0]), '+k')
    plt.plot(mi_value, np.zeros(mi_value.size), 'r*')
    plt.ylim([-(ymax*0.1),ymax*1.1])
    plt.show()
```

### Scoring Kernel Density Estimate

```python
def KDEScoring(values, bandwidth):
    kde = KernelDensity(kernel='gaussian', bandwidth=bandwidth).fit(values)
    return kde.score(values)
```
