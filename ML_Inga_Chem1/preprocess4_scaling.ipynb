```python
import numpy as np
import pandas as pd

pd.set_option('max_columns', 100)
```

```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10,5]
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

### Reading .csv file

```python
raw = pd.read_csv('training_dropped.csv')
raw.head()
```

```python
raw.shape
```

### Extracting features, labels, and confidence score

```python
feature = raw.iloc[:,:6215]
feature.head()
```

```python
label = raw.iloc[:,6215:-1]
label.head()
```

```python
confidence = raw.iloc[:,-1]
confidence.head()
```

### Exacting fragment and loss features

```python
loss = feature.loc[:,feature.columns.str.match('loss')]
loss.head()
```

```python
fragment = feature.loc[:,feature.columns.str.match('fragment')]
fragment.head()
```

### Number unique values

```python
fragment.apply(lambda x: ','.join(str(i) for i in x.unique()), axis=0)
```

```python
loss.apply(lambda x: ','.join(str(i) for i in x.unique()), axis=0)
```

### Scaling fragment features and keeping loss features remain unchanged

```python
mas = MaxAbsScaler()
```

```python
mat = mas.fit_transform(fragment)
mat.shape
```

```python
np.apply_along_axis(lambda x: ','.join(str(i) for i in np.unique(x)), 0, mat)
```

```python
newfragment = pd.DataFrame(mat, columns=fragment.columns)
newfragment.head()
```

### Write Data to CSV File

```python
newraw = pd.concat([newfragment, loss, label, confidence], axis=1)
newraw.head()
```

```python
newraw.shape
```

```python
newraw.to_csv('training_scaled.csv', index=None)
```
